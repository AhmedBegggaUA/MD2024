```python
from matplotlib import rcParams, cycler
import matplotlib.pyplot as plt
import numpy as np
plt.ion()
```

```python
# Fixing random state for reproducibility
np.random.seed(19680801)

N = 10
data = [np.logspace(0, 1, 100) + np.random.randn(100) + ii for ii in range(N)]
data = np.array(data).T
cmap = plt.cm.coolwarm
rcParams['axes.prop_cycle'] = cycler(color=cmap(np.linspace(0, 1, N)))


from matplotlib.lines import Line2D
custom_lines = [Line2D([0], [0], color=cmap(0.), lw=4),
                Line2D([0], [0], color=cmap(.5), lw=4),
                Line2D([0], [0], color=cmap(1.), lw=4)]

fig, ax = plt.subplots(figsize=(10, 5))
lines = ax.plot(data)
ax.legend(custom_lines, ['Cold', 'Medium', 'Hot']);
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 4, 5, 6]

fig, ax = plt.subplots(1, 2, figsize=(10, 5)) # Create a figure and 2 axes. Note that figsize is the size of the figure
ax[0].plot(x, y, label='Line 1') #Â Plot in the first axis, and add a label
ax[0].set_title('Line Plot') #Â We can also add a title to the plot
#Â We can also add labels to the x-axis and y-axis
ax[0].set_xlabel('X-axis')
ax[0].set_ylabel('Y-axis')
#Â We can also change the color, the line style, and the marker style of the plot
ax[0].plot(y, x, color='red', linestyle='dashdot', marker='x', label='Line 2')
#Â We can also add a legend to the plot
ax[0].legend(['Line 1', 'Line 2'])

ax[1].scatter(x, y, label='Scatter 1') #Â Plot in the second axis
ax[1].set_title('Scatter Plot')
ax[1].set_xlabel('X-axis')
ax[1].set_ylabel('Y-axis')
ax[1].scatter(y, x, color='green', marker='o', label='Scatter 2')
ax[1].legend(['Scatter 1', 'Scatter 2'])
plt.savefig('./images/practices/multiple2.png')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 4, 5, 6]

fig, ax = plt.subplots(1, 2, figsize=(10, 5)) # Create a figure and 2 axes. Note that figsize is the size of the figure
sns.lineplot(x=x, y=y, label='Line 1', color='purple', linestyle='--', ax=ax[0]) #Â Plot in the first axis, and add a label
ax[0].set_title('Line Plot') #Â We can also add a title to the plot
#Â We can also add labels to the x-axis and y-axis
ax[0].set_xlabel('X-axis')
ax[0].set_ylabel('Y-axis')
#Â We can also change the color, the line style, and the marker style of the plot
sns.lineplot(x=y, y=x, color='black', linestyle='dashdot', ax=ax[0], label='Line 2')
#Â We can also add a legend to the plot
ax[0].legend(['Line 1', 'Line 2'])

sns.scatterplot(x=x, y=y, color='green', marker='o', ax=ax[1], label='Scatter 1') #Â Plot in the second axis
ax[1].set_title('Scatter Plot')
ax[1].set_xlabel('X-axis')
ax[1].set_ylabel('Y-axis')
sns.scatterplot(x=y, y=x, color='orange', marker='x', ax=ax[1], label='Scatter 2')
ax[1].legend(['Scatter 1', 'Scatter 2'])
plt.savefig('./images/practices/seaborn5.png')
plt.show()
```
